```python
tokens = []
labels = []
with open('./output/result_dir/token_test.txt', 'r') as f:
    for line in f.readlines():
        tokens.append(line)
with open('./output/result_dir/label_test.txt', 'r') as f:
    for line in f.readlines():
        labels.append(line)
```
